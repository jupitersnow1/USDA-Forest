## Question 3

 extract the total amount of the donations from Indiana during an election year

```r
# code here
myindydonations <- function(myyear) {
    myDF <- fread(paste0("/anvil/projects/tdm/data/election/itcont", myyear, ".txt"), quote="", select = c(10,15))
    names(myDF) <- c("state", "donation")
    myvalue <- tapply(myDF$donation, myDF$state, sum)['IN']
    names(myvalue) <- myyear
    return(myvalue)
}
```

test this function by discovering how much money was donated from Indiana during the 1990 election cycle:

```r
library(data.table)
myindydonations(1990)
```

sapply function to run this function on each election year (in other words, the even numbered years) from 1980 to 2018

```r
library(data.table)
myresults <- sapply( seq(1980,2018,by=2), myindydonations )
```

which yields the total amount of money donated from Indiana during each election cycle from 1980 to 2018

```r
plot(names(myresults), myresults)
```

Q: Plot amount of money donated from Indiana per election cycle from 1980 to 2018.

first extract the total amount of the taxi cab rides per day of a given month

```r
myfares <- function(mymonth) {
    myDF <- fread(paste0("/anvil/projects/tdm/data/taxi/yellow/yellow_tripdata_2018-", mymonth, ".csv"), select=c(2,17))
    mytable <- tapply(myDF$total_amount, as.Date(myDF$tpep_pickup_datetime), sum)
    return(mytable)
}
```

test this function by discovering how much money was spent on each day in January

```r
library(data.table)
myfares("01")
```

use the sapply function to run this function on each month from "01" to "12"

```r
# yields the total amount of money spent on taxi cab rides each day.
library(data.table)
myresults <- sapply( sprintf("%02d", 1:12), myfares )
```

```r
names(myresults) <- NULL
v <- do.call(c, myresults)
mytotals <- tapply(v, names(v), sum)
betterdates <- mytotals[year(as.Date(names(mytotals))) == 2018]
```

### Example 1 <br>
we will be using the 1980s election data to find the amount of money donated in each state. 

Functions used: <br>
`fread()`: <br>
In this example we are using fread() instead of read.csv because the column names are not included in the dataset and also becuase the data was not separated by commas <br> <br>
`tapply([vect], [split by], [func])`: <br> 
we are going to add up using the sum (func)tion the transaction amount(vect) to election campaigns for each state (vect). 

```r
library(data.table)
myDF <- fread("/anvil/projects/tdm/data/election/itcont1980.txt", quote = "")
names(myDF) <- c("CMTE_ID", "AMNDT_IND", "RPT_TP", "TRANSACTION_PGI", "IMAGE_NUM", "TRANSACTION_TP", "ENTITY_TP", "NAME", "CITY", "STATE","ZIP_CODE", "EMPLOYER", "OCCUPATION", "TRANSACTION_DT", "TRANSACTION_AMT", "OTHER_ID", "TRAN_ID", "FILE_NUM", "MEMO_CD", "MEMO_TEXT", "SUB_ID")
```

```r
# this is adding the total transact_amt of each individual state
tapply(myDF$TRANSACTION_AMT, myDF$STATE, sum) 
```
